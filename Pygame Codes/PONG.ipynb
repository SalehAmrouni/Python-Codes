```python
#imports
import pygame
from random import choice
from random import *
```

```python
#frames
fps =80
```

```python
#colors
white = (255,255,255)
red = (255,0,0)
black = (0,0,0)
```

```python
#ball coordinates
x_ball = 500
y_ball = 250
```

```python
#the score of the player
score = 0
```

```python
#where position is updated
step = 2
```

```python
#clock to control fps
clock = pygame.time.Clock()
```

```python
#directions
horizontal = 'left'
vertical = 'down'
```

```python
#display
screen = pygame.display.set_mode((500,500))
```

```python
#initalize pygame display
pygame.init()
```

```python
#draw the ball
def ball():
    pygame.draw.circle(screen, red,(x_ball,y_ball),10,10)
```

```python
#draw the paddle
def paddle(x,y):
    pygame.draw.rect(screen, black,(x,y,80,10))
```

```python
#move the ball
def move_ball():
    global x_ball, y_ball, horizontal, vertical
    if horizontal == "left":
        x_ball -= step
     #movments for the ball   
        if x_ball < 10:
            side_side = "right"
    if vertical == "down":
        y_ball += step
    if vertical == "up":
        y_ball -= step
    if vertical == 'up':
        y_ball -= step
        if y_ball < 10:
            up_down = 'down'
    if horizontal == 'right':
        x_ball += step
    if x_ball > 480:
        up_down = 'left'
```

```python
#what to do if hit
def served():
    #making variable global for use out of functions
    global x_ball, y_ball
    global x,y
    global side_side, up_down
    global score
    #movments fpr when hit
    if horizontal == 'left':
        if y_ball > 480:
            
            if x_ball >= x and x_ball < x + 50:
                print('good job')
                
                score += 1
                
                vertical = 'up'
                step = 1
                #what to do if missed
            else:
                pygame.draw.circle(screen,black,(x_ball, y_ball),  10,10)
                pygame.display.update()
                x_ball, y_ball = 500, 250
                score -= 1
                
        pygame.display.set_caption(f'your score is: {score}')
```

```python
#infinte loop for the game
loop = True
while loop:
    keys = pygame.key.get_pressed()
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            loop = 0 
        x,y = pygame.mouse.get_pos()
    #functions that we made are being declared including variables
    move_ball()
    ball()
    paddle(x,490)
    served()
    pygame.display.update()
    
    screen.fill(white)
    
    clock.tick(fps)
pygame.quit()
```
